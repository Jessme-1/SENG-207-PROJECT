## MEAN OF SPECIFIC COLUMNS

```python
meanFA=dataset.fixed_acidity.mean()
meanCc=dataset.citric_acid.mean()
meanCh=dataset.chlorides.mean()
meanFso=dataset.freesulfurdioxide.mean()
meanTso=dataset.totalsulfurdioxide.mean()
meanD=dataset.density.mean()
meanpH=dataset.pH.mean()
meanquality=dataset.quality.mean()
```

```python
print(meanFA)
print(meanCc)
print(meanCh)
print(meanFso)
print(meanTso)
print(meanD)
print(meanpH)
print(meanquality)
```

## FILLING IN NULL DATAS WITH THEIR MEAN

```python
dataset['fixeda_cidity'] = dataset['fixed_acidity'].fillna(meanFA)
dataset['citri_cacid'] = dataset['citric_acid'].fillna(meanCc)
dataset['chlorides'] = dataset['chlorides'].fillna(meanCh)
dataset['freesulfurdioxide'] = dataset['freesulfurdioxide'].fillna(meanFso)
dataset['totalsulfurdioxide'] = dataset['totalsulfurdioxide'].fillna(meanTso)
dataset['density'] = dataset['density'].fillna(meanD)
dataset['pH'] = dataset['pH'].fillna(meanpH)
dataset['quality'] = dataset['quality'].fillna(meanquality)
```

## PIE CHART

```python
meanOf_dataset=np.array([8.319875,15.886875,46.52,3.31110625,5.63625])
myLabels=['fixed_acidity','freeSO','totalSO','pH','quality',]
explode=(0,0.2,0,0.2,0,)

plt.figure(figsize=(6,9))
plt.title('Mean Of specific columns', fontsize=8, fontweight='bold')
plt.style.use('ggplot')
plt.pie(meanOf_dataset,labels=myLabels,explode=explode,startangle=0,autopct='%.1f%%')
plt.legend(loc='upper left', fontsize=6)
plt.show()
```

The pie chart describes the columns according to their averaged values, wwhich indicates the quantity of each columns in the beverage. The quality is low for the averaged value.
